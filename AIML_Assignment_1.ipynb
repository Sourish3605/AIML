```python
import numpy as np
arr = np.array([1, 2, 3, 4, 5])
sin_values = np.sin(arr)
log_values = np.log(arr)
log2_values = np.log2(arr)
exp_values = np.exp(arr)

print("Sine values:", sin_values)
print("Natural log values:", log_values)
print("Base-2 log values:", log2_values)
print("Exponential values:", exp_values)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0.1, 10, 100)
y_sin = np.sin(x)
y_log = np.log(x)
y_exp = np.exp(x)

plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
plt.plot(x, y_sin, label='sin(x)')
plt.title('Sine Function')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(x, y_log, label='log(x)', color='orange')
plt.title('Logarithm Function')
plt.xlabel('x')
plt.ylabel('log(x)')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(x, y_exp, label='exp(x)', color='green')
plt.title('Exponential Function')
plt.xlabel('x')
plt.ylabel('exp(x)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
people = ['Kiran', 'Arun', 'Vijay', 'Varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
data = {
    'Name': people,
    'Age': age,
    'Height': height,
    'Weight': weight
}
df = pd.DataFrame(data)
print(df)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Age'], df['Weight'], color='blue', marker='o')
plt.title('Scatter Plot of Age vs Weight')
plt.xlabel('Age (years)')
plt.ylabel('Weight (kg)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df['Name'], df['Height'], color='green')
plt.title('Bar Chart of Individuals\' Heights')
plt.xlabel('Name')
plt.ylabel('Height (cm)')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Age'], bins=5, color='orange', edgecolor='black')
plt.title('Histogram of Age Distribution')
plt.xlabel('Age (years)')
plt.ylabel('Frequency')
plt.show()
```
